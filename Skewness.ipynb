## Skewness 
* Skewness measures the asymmetry of a data distribution how far it deviates from being perfectly balanced.
* In a symmetric dataset, the left and right sides of the graph mirror each other (Mean Median = Mode).
* When that balance breaks, skewness tells us which side the data tail stretches toward.
* It helps understand the data shape, identify outliers, and choose the right statistical methods for analysis.

## Right Skewed

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Right skewed data
right_skew = np.random.exponential(scale=2, size=1000)

plt.figure(figsize=(7,4))
sns.histplot(right_skew, kde=True)

plt.title("Right Skew (Positive Skew)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

right_skew = np.random.exponential(scale=2, size=1000)

plt.figure(figsize=(7,4))
sns.histplot(right_skew, kde=True)

plt.axvline(np.mean(right_skew), color='red', linestyle='--', label=f"Mean = {np.mean(right_skew):.2f}")
plt.axvline(np.median(right_skew), color='green', linestyle='-', label=f"Median = {np.median(right_skew):.2f}")

plt.title("Right Skew (Positive Skew)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

## Normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Normal distribution
normal_data = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(7,4))
sns.histplot(normal_data, kde=True)

plt.title("Normal Distribution (No Skew)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Example: your DataFrame
# Replace with your own df
df = pd.DataFrame({
    'A': [1, 2, 3, 40],
    'B': [10, 10, 9, 8],
    'C': [5, 6, 7, 8]
})

# Loop through numeric columns
for col in df.select_dtypes(include='number'):
    data = df[col].dropna()

    plt.figure(figsize=(7,4))
    sns.histplot(data, kde=True)

    # mean & median lines
    plt.axvline(data.mean(), color='red', linestyle='--', label=f"Mean = {data.mean():.2f}")
    plt.axvline(data.median(), color='green', linestyle='-', label=f"Median = {data.median():.2f}")

    plt.title(f"Skewness Plot of '{col}' (Skew = {data.skew():.2f})")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()
```
