## Measures of Dispersion:
Dispersion is the state of getting
dispersed or spread. Statistical dispersion means the extent to which
numerical data is likely to vary about an average value. In other words,
dispersion helps to understand the distribution of the data.


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
```

## 1. Range
* The range describes the variability of the dataset by subtracting the highest value from the lowest value in the data.
* Range give quick understanding of the overall spread of data but sensitive to outliers
* Example: 60,70, 75, 80, 100 Range=100-60=40

```python
min_r = df["age"].min()
max_r = df["age"].max()
```

```python
min_r,max_r
```

```python
range_ = max_r - min_r
range_
```

## 2. Mean Absolute Deviation
 
   Mean Absolute Deviation (MAD) measures how much each data point deviates (differs) from the mean of the dataset  on average.

   * A low MAD  data points are close to the mean  less variation.

   * A high MAD  data points are widely spread  more variation.

```python
sec_a =np.array ([75, 65, 73, 68, 72, 76])
sec_b =np.array ([90, 47, 43, 96, 93, 51])
no = np.array ([1,2,3,4,5,6])
```

```python
mean = np.mean(sec_b)
print(mean)
```

Formula of mean_absolute_deviation

```python
mean_absolute_deviation_a = np.sum(np.abs(sec_a - mean))/len(sec_a)
mean_absolute_deviation_b = np.sum(np.abs(sec_b - mean))/len(sec_b)
print(mean_absolute_deviation_a)
print(mean_absolute_deviation_b)
```

```python
plt.figure(figsize=(10,3))
plt.scatter(sec_a,no,label="sec A")
plt.scatter(sec_b,no,color="red",label="sec B")
plt.plot([70,70,70,70,70,70],no,c="blue",label="mean")
plt.legend()
plt.show()
```

```python
np.std(sec_a) , np.std(sec_b)
```
