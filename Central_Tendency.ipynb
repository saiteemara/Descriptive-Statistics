```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head(3)
```

## Measures of Central Tendency

1. Mean

```python
df["age"].mean()
```

```python
np.mean(df["age"])
```

```python
sns.histplot(x="age",data=df)
plt.show()
```

```python
sns.histplot(x="age",data=df,bins=[i for i in range(0,81,10)])
plt.show()
```

```python
mn = np.mean(df["age"])
```

```python
sns.histplot(x="age",data=df,bins=[i for i in range(0,81,10)])
plt.plot([mn for i in range(0,300)],[i for i in range(0,300)],c="red")
plt.show()
```

2. Median

```python
np.median(df["age"])
```

```python
df.isnull().sum()
```

```python
df["age"].fillna(df["age"].mean(),inplace=True)
```

```python
np.median(df["age"])
```

```python
md_fare = np.median(df["fare"])
md_fare
```

```python
mn_fare = np.mean(df["fare"])
mn_fare
```

```python
sns.histplot(x="fare",data=df,bins=[i for i in range(0,81,10)])
plt.plot([mn_fare for i in range(0,300)],[i for i in range(0,300)],c="red")
plt.plot([md_fare for i in range(0,300)],[i for i in range(0,300)],c="blue")
plt.show()
```

3. Mode

```python
df["fare"].mode()
```

```python
df["fare"].value_counts()
```

```python
mo_fare = df["fare"].mode()[0]
mo_fare
```

```python
sns.histplot(x="fare",data=df,bins=[i for i in range(0,81,10)])
plt.plot([mn_fare for i in range(0,300)],[i for i in range(0,300)],c="green",label="mean")
plt.plot([md_fare for i in range(0,300)],[i for i in range(0,300)],c="blue",label="medain")
plt.plot([mo_fare for i in range(0,300)],[i for i in range(0,300)],c="red",label="mode")
plt.legend()
plt.show()
```
